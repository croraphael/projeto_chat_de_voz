#### Código eficiente para o sistema do próprio computador

```python
import sounddevice as sd
from scipy.io.wavfile import write
import os

def record(sec=5, file_name='request_audio.wav'):
    fs = 44100  # Frequência de amostragem (qualidade padrão)
    print(f"Gravando por {sec} segundos...")
    
    # Grava o áudio do microfone
    # channels=1 para mono, channels=2 para estéreo
    myrecording = sd.rec(int(sec * fs), samplerate=fs, channels=1)
    
    sd.wait()  # Espera a gravação terminar
    print("Gravação finalizada!")
    
    # Salva o arquivo no diretório atual
    write(file_name, fs, myrecording)
    
    # Retorna o caminho completo do arquivo
    return os.path.abspath(file_name)

# Uso
record_file = record(5)

# Se estiver no Jupyter do VS Code, o display(Audio) ainda funciona:
from IPython.display import Audio, display
display(Audio(record_file, autoplay=True))
```
